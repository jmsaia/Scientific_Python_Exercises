## Requirement 21

```python
# Demonstrate the use of five Pandas str methods on a Pandas series named
# states created from a list of seven names of U.S. states.

state_list = ['Louisiana', 'Hawaii', 'Texas', 'New York', 'Illinois', 'California', 'Oregon']

states =pd.Series(state_list)

print(states.str.len(), '\n')
print(states.str.swapcase(), '\n')
print(states.str.lower(), '\n')
print(states.str.startswith('L'), '\n')
print(states.str.endswith('i'))
```

## Requirement 22

```python
# Demonstrate the use of three regular expression examples using the states
# Pandas series.

print(states.str.extract('([A-Za-z]+)', expand=False), '\n')
print(states.str.findall(r'^[^AEIOU].*[^aeiou]$'), '\n')
print(states.str.count('[A-Za-z]'))
```

## Requirement 23

```python
# Demonstrate the use of three miscellaneous methods using the states
# Pandas series. 

print(states.str.slice(0,5), '\n')
print(states.str.cat(), '\n')
print(states.str.join('_'))
```

```python
# Use the timeit function to compare processes summing four Pandas dataframe objects using:
# the typical approach (simple + operations)

column_name = ['A','B','C','D']

data1 = np.random.randint(10, size = (4,4))
data2 = np.random.randint(10, size = (4,4))
data3 = np.random.randint(10, size = (4,4))
data4 = np.random.randint(10, size = (4,4))

df1_4 = pd.DataFrame(data1, columns=column_name)
df2_4 = pd.DataFrame(data2, columns=column_name)
df3_4 = pd.DataFrame(data3, columns=column_name)
df4_4 = pd.DataFrame(data4, columns=column_name)

                           
%timeit df1_4 + df2_4 + df3_4 + df4_4

# the pd.eval() approach

%timeit pd.eval('df1_4 + df2_4 + df3_4 + df4_4')
```
